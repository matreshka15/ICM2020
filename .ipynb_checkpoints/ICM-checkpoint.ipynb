## 数据可视化
* 函数：plot_event()
    * 参数：
        * match_record - Pandas数据记录中的一条（一行）
        * line_index - 要显示第几行数据的可视化
        * draw_line - 是否需要画出球的运动轨迹
        * grid - 网格化后每个格子的长x、宽y
    * 返回值： 一个plot
* 函数：Coordinates_convert()
    * 参数：piece_record,一行Events记录
    * 返回值：转换后的坐标，始终以Hurskies在左边，Opponents在右边为参考系。

```python
def Coordinates_convert(piece_record):
    if 'Huskies' in str(piece_record.OriginPlayerID):
        Ocoordinates = (piece_record.EventOrigin_x,piece_record.EventOrigin_y)
        Dcoordinates = ((piece_record.EventDestination_x),(piece_record.EventDestination_y))
    else:
        Ocoordinates = ((100-piece_record.EventOrigin_x),(100-piece_record.EventOrigin_y))
        Dcoordinates = ((100-piece_record.EventDestination_x),(100-piece_record.EventDestination_y))

    if (matches.loc[matches.MatchID == piece_record.MatchID].Side.tolist() == ['home']):
        if(str('Huskies') in piece_record.OriginPlayerID):
            if(Ocoordinates==(0,0)):
                Ocoordinates = (0,50)
            if(Dcoordinates==(0,0)):
                Dcoordinates = (100,50)
            elif(Dcoordinates==(100,100)):
                Dcoordinates = (0,50)
        else:
            if(Ocoordinates==(100,100)):
                Ocoordinates = (100,50)
            if(Dcoordinates==(0,0)):
                Dcoordinates = (0,50)
            elif(Dcoordinates==(100,100)):
                Dcoordinates = (100,50)
    else:
        if(str('Huskies') in piece_record.OriginPlayerID):
            if(Ocoordinates==(100,100)):
                Ocoordinates = (0,50)
            if(Dcoordinates==(0,0)):
                Dcoordinates = (0,50)
            elif(Dcoordinates==(100,100)):
                Dcoordinates = (100,50)
        else:
            if(Ocoordinates==(0,0)):
                Ocoordinates = (100,50)
            if(Dcoordinates==(0,0)):
                Dcoordinates = (100,50)
            elif(Dcoordinates==(100,100)):
                Dcoordinates = (0,50) 

    Ocoordinates = (Ocoordinates[0],100-Ocoordinates[1])
    Dcoordinates = (Dcoordinates[0],100-Dcoordinates[1])
    return Ocoordinates,Dcoordinates
```

```python
def plot_event(match_record,line_index,draw_line = True,grid=(0,0)):
    
    file_path = os.path.join(r'Misc/','background.png')
    background = imgplt.imread(file_path)
    width = (10,6)
    plt.figure(figsize=width)
    plt.imshow(background,aspect='auto',extent=(0,100,0,100))
    for index in line_index:        
        Olegend = match_record.iloc[index].OriginPlayerID
        Dlegend = match_record.iloc[index].DestinationPlayerID
        if 'Huskies' in str(match_record.iloc[index].DestinationPlayerID):
            Dcolor = 'b'
        elif 'Oppo' in str(match_record.iloc[index].DestinationPlayerID):
            Dcolor = 'r'
        else:
            Dcolor = 'c'
            Dlegend = 'nobody'
        if 'Huskies' in str(match_record.iloc[index].OriginPlayerID):
            Ocolor = 'b'
        else:
            Ocolor = 'r'
        Ocoordinates,Dcoordinates = Coordinates_convert(match_record.iloc[index])
        
        EventType = str(encoders['EventType'].inverse_transform([match_record.iloc[index].EventType])[0] )
        EventSubType = str(encoders['EventSubType'].inverse_transform([match_record.iloc[index].EventSubType])[0])
        if draw_line:
            plt.arrow(Ocoordinates[0],Ocoordinates[1],Dcoordinates[0]-Ocoordinates[0],Dcoordinates[1]-Ocoordinates[1],\
                      fill=False,ec='white',width=0.001,length_includes_head=True,head_width=2, head_length=2)
        Origin = plt.scatter(Ocoordinates[0],Ocoordinates[1],c=Ocolor)
        Destination = plt.scatter(Dcoordinates[0],Dcoordinates[1],c=Dcolor)
        plt.annotate(index, xy =(Ocoordinates[0],Ocoordinates[1]), xytext = (Ocoordinates[0]+0.1,Ocoordinates[1]+0.1))
    if(len(line_index)==1): 
        plt.legend([Origin,Destination],['Origin:  '+Olegend,'Destination:  '+Dlegend])
        plt.title(EventType+'->'+EventSubType)
        
    #plot grid
    if(0 not in grid):
        x_span = int(100/grid[0])
        y_span = int(100/grid[1])
        for cnt in range(x_span):
            plt.vlines(cnt*grid[0], 0, 100,color="brown",linestyles='dashed')
        for cnt in range(y_span):
            plt.hlines(cnt*grid[1], 0, 100,color="brown",linestyles='dashed')
    ax = plt.gca()
    ax.set_xlim(-1, 101)
    ax.set_ylim(-1, 101)
```

### 生成网络
* 函数：
    * weight_transformation():在数据集中每见到一对元素出现，对权重进行的操作。
        * 例如：每次+1；或每次指数增长等。
    * build_network():生成网络。
        * CoDataFrame - 包含网络中每两个元素交互的DataFrame 
        * node_size - 生成图中的最大节点大小
        * factor - 网络图中边与权重的比例系数
        * initial_value - 每条边初始化时的值

```python
huskies_passing_events = passing_events.loc[passing_events.TeamID=='Huskies'].reset_index(drop=True)
G_max_MatchID = huskies_passing_events.MatchID.max() # 比赛最大场数
G_Match_dict = {} # 字典里存放各[场]比赛的passes数据
for match_index in range(1,G_max_MatchID+1):
    G_Match_dict['Match'+ str(match_index)] = huskies_passing_events.loc[huskies_passing_events.MatchID == match_index].reset_index(drop=True)
```

```python
def weight_transform(original_edge_weight):
    original_edge_weight += 1
    return original_edge_weight

def build_network(CoDataFrame,pos={},extract_list=[],node_size=1500,factor=3,threshold=0,initial_value=1,font_size=10,new_figure=True,plot = True):
    file_path = os.path.join(r'Misc/','background (2).png')
    background = imgplt.imread(file_path)
    
    element_set = []
    for col in CoDataFrame.columns:
        element_set += CoDataFrame[col].tolist()
    element_set = set(element_set)
    #print('节点数：{}'.format(len(element_set)))
    G = nx.DiGraph()
    G.add_nodes_from(element_set)
    
    for index in range(CoDataFrame.index.max()+1):
        temp = sorted(CoDataFrame.iloc[index].tolist())
        if((temp[0],temp[1]) not in set(G.edges)):
            G.add_edge(temp[0],temp[1],weight=initial_value)
            if((temp[0],temp[1]) in list(G.edges) == False):
                print(temp,'++',(temp[0],temp[1]) in set(G.edges))
        else:
            G[temp[0]][temp[1]]['weight'] = weight_transform(G[temp[0]][temp[1]]['weight'])
    if(new_figure):
        plt.figure(figsize=(10,6))
    cut_branch = [(u,v) for (u,v,d) in G.edges(data=True)if d['weight']<threshold ]
    G.remove_edges_from(cut_branch)
    G.remove_nodes_from([k for k,v in G.degree() if v == 0])
    if(len(extract_list) != 0):
        extract_nodes = list(set(G.nodes) - set(extract_list))  
        G.remove_nodes_from(extract_nodes)
    if(len(pos.keys())==0):
        pos=nx.spring_layout(G,iterations=20)
    if(plot):
        nx.draw_networkx_nodes(G,pos,node_size=node_size,node_color='black')
        nx.draw_networkx_edges(G,pos,width=[float(d['weight']*factor) for (u,v,d) in G.edges(data=True)],alpha=0.6,edge_color='blue')
        #nx.draw_networkx_labels(G,pos,font_size=font_size,font_family='sans-serif',font_color='white')
        plt.imshow(background,aspect='auto',extent=(0,100,0,100))
        #plt.show()
    return G
```

## 匹配三元组
* url: https://stackoverflow.com/questions/55339231/get-the-list-of-triad-nodes-who-fall-under-the-category-of-individual-triadic
* 算法说明：请看论文。。。

```python
import itertools
def _tricode(G, v, u, w):
    """Returns the integer code of the given triad.

    This is some fancy magic that comes from Batagelj and Mrvar's paper. It
    treats each edge joining a pair of `v`, `u`, and `w` as a bit in
    the binary representation of an integer.

    """
    combos = ((v, u, 1), (u, v, 2), (v, w, 4), (w, v, 8), (u, w, 16),
              (w, u, 32))
    return sum(x for u, v, x in combos if v in G[u])

#G_origin = G
#G = G.to_directed()
def triads_analysis(G):
    #: The integer codes representing each type of triad.
    #: Triads that are the same up to symmetry have the same code.
    TRICODES = (1, 2, 2, 3, 2, 4, 6, 8, 2, 6, 5, 7, 3, 8, 7, 11, 2, 6, 4, 8, 5, 9,
                9, 13, 6, 10, 9, 14, 7, 14, 12, 15, 2, 5, 6, 7, 6, 9, 10, 14, 4, 9,
                9, 12, 8, 13, 14, 15, 3, 7, 8, 11, 7, 12, 14, 15, 8, 14, 13, 15,
                11, 15, 15, 16)

    #: The names of each type of triad. The order of the elements is
    #: important: it corresponds to the tricodes given in :data:`TRICODES`.
    TRIAD_NAMES = ('003', '012', '102', '021D', '021U', '021C', '111D', '111U',
                   '030T', '030C', '201', '120D', '120U', '120C', '210', '300')

    #: A dictionary mapping triad code to triad name.
    TRICODE_TO_NAME = {i: TRIAD_NAMES[code - 1] for i, code in enumerate(TRICODES)}

    triad_nodes = {name: set([]) for name in TRIAD_NAMES}
    m = {v: i for i, v in enumerate(G)}
    for v in G:
        vnbrs = set(G.pred[v]) | set(G.succ[v])
        for u in vnbrs:
            if m[u] > m[v]:
                unbrs = set(G.pred[u]) | set(G.succ[u])
                neighbors = (vnbrs | unbrs) - {u, v}
                not_neighbors = set(G.nodes()) - neighbors - {u, v}
                # Find dyadic triads
                for w in not_neighbors:
                    if v in G[u] and u in G[v]:
                        triad_nodes['102'].add(tuple(sorted([u, v, w])))
                    else:
                        triad_nodes['012'].add(tuple(sorted([u, v, w])))
                for w in neighbors:
                    if m[u] < m[w] or (m[v] < m[w] < m[u] and
                                       v not in G.pred[w] and
                                       v not in G.succ[w]):
                        code = _tricode(G, v, u, w)
                        triad_nodes[TRICODE_TO_NAME[code]].add(
                            tuple(sorted([u, v, w])))
    # find null triads
    all_tuples = set()
    for s in triad_nodes.values():
        all_tuples = all_tuples.union(s)
    triad_nodes['003'] = set(itertools.combinations(G.nodes(), 3)).difference(all_tuples)
    return triad_nodes

triad_nodes = triads_analysis(G)
```

## 长短期结构关系分析
* 通过设置时间窗口来过滤掉‘突发性’事件

```python
def Generate_slices(minutes=10,Match = 'Match1',MatchPeriod = '1H'):
    TimeSlice = 60*minutes  # 10min time window
    TimeGap = TimeSlice/10 
    G_Match_Window = G_Match_dict[Match].loc[G_Match_dict[Match].MatchPeriod==MatchPeriod]
    Match_Slices = []
    cnt = 1
    while len(G_Match_Window) > 0:   
        temp = G_Match_Window.loc[G_Match_Window.EventTime <= cnt*TimeSlice]
        Match_Slices.append(temp)
        G_Match_Window = G_Match_Window.loc[G_Match_Window.EventTime > cnt*(TimeSlice-TimeGap)]
        cnt += 1
    time_window_structure = []
    for sub_time_slice_index in range(len(Match_Slices)):
        sub_time_slice = Match_Slices[sub_time_slice_index].reset_index(drop=True)
        CoDataFrame_slice = sub_time_slice[['OriginPlayerID','DestinationPlayerID']]
        G_slice = build_network(CoDataFrame_slice,initial_value=1,new_figure=False,plot=False)
        time_window_structure.append(triads_analysis(G_slice.to_directed()))
    print('共产生{}个时间片数据'.format(len(time_window_structure))) 
    return time_window_structure
```

```python
def compare_combinations(common,compare):
    popup_cnt = 0
    for key in common.keys():
        combination = list(common[key])
        temp = combination
        for everycombination in combination:
            if(key in compare.keys()):
                if everycombination not in compare[key]:
                    temp.remove(everycombination)
                    #print('removing ',everycombination,'in ',key)
                    popup_cnt += 1
        common[key] = temp
    return common,popup_cnt
```

#### 对比整个半场比赛中以时间片TimeSlice划分的所有小片段，筛选出每次都会在其中出现的结构

```python
def filter_time_window(time_window_structure):
    common = time_window_structure[0]
    popup_cnt = 0
    total_pops = 0
    for item in range(len(time_window_structure)-1):
        compare = time_window_structure[item+1]
        common,popup_cnt = compare_combinations(common,compare)
        total_pops += popup_cnt
    print('共弹出{}条事件'.format(total_pops))
    return total_pops,common
```

```python
MATCH_NUMBER = len(G_Match_dict)
all_matches_G = {}
repeater = '1H'
for half in range(MATCH_NUMBER):
    full_match = str(half+1)
    for coin in range(2):
        CoDataFrame = G_Match_dict['Match'+ full_match].loc[G_Match_dict['Match'+full_match].MatchPeriod==repeater][['OriginPlayerID','DestinationPlayerID']]
        location = G_Match_dict['Match'+ full_match].loc[G_Match_dict['Match'+full_match].MatchPeriod==repeater][['OriginPlayerID','DestinationPlayerID','EventDestination_x','EventDestination_y','EventOrigin_x','EventOrigin_y']].reset_index(drop=True)
        for item in set(location.OriginPlayerID.unique().tolist()+location.DestinationPlayerID.unique().tolist()):
            Oxy = location.loc[location.OriginPlayerID == item][['EventOrigin_x','EventOrigin_y']]
            Dxy = location.loc[location.DestinationPlayerID == item][['EventDestination_x','EventDestination_y']]
            if(len(Oxy) == 0):
                Oxmean = 0
                Oymean = 0
            else:
                Oxmean = Oxy.EventOrigin_x.mean()
                Oymean = Oxy.EventOrigin_y.mean()
            if(len(Dxy) == 0):
                Dxmean = 0
                Dymean = 0
            else:
                Dxmean = Dxy.EventDestination_x.mean()
                Dymean = Dxy.EventDestination_y.mean()   
            x =  round(Oxmean+Dxmean)/2
            y = round(Dxmean+Dymean)/2

            pos[item]=(x,y)
        CoDataFrame = CoDataFrame.reset_index(drop=True)
        G = build_network(CoDataFrame,pos=pos,node_size=1000,factor=1,threshold=0,initial_value=1,new_figure=False,plot=False)
        all_matches_G['Match'+full_match+repeater] = G
        if(repeater == '1H'):
            repeater = '2H'
        else:
            repeater = '1H'
```

```python
indicator = data_to_eval.pops/(data_to_eval.triad003+data_to_eval.triad102+data_to_eval.triad201+data_to_eval.triad300)


mental_coefficient = 0.5
tactical_coefficient = 1-mental_coefficient



def snow_white_indicator(data_to_eval,mental_coefficient=0.5):
    mental_coefficient = 0.5
    tactical_coefficient = 1-mental_coefficient
    structure_factor = (-0.4*0*data_to_eval.triad102+data_to_eval.triad003-0.4*0*data_to_eval.triad201+data_to_eval.triad300)\
                        /data_to_eval.total_triads
    mental_factor = (tactical_coefficient*data_to_eval.passing_ratio+mental_coefficient*data_to_eval.side)
    tatical_factor = tactical_coefficient*(structure_factor)#*(1/structure_factor.std()))/(np.max(1/structure_factor.std()))
    confusion_factor = (tactical_coefficient/2)*(data_to_eval.pops/data_to_eval.total_triads)
    dynamic_factor = data_to_eval.huskies_action_ratio
    indicator_evaluate_team = 0.5*mental_factor+tatical_factor+0.4*dynamic_factor -confusion_factor
    return indicator_evaluate_team,confusion_factor,tatical_factor,mental_factor,dynamic_factor

indicator_evaluate_team,_,_,_,_ = snow_white_indicator(data_to_eval,mental_coefficient=0.2)

plt.figure(figsize=(9,4))
plt.title('SW indicator versus simplified H indicator')
ax = plt.subplot(121)
ax.set_yticks([0,1])
target = data_to_eval.outcome
sns.kdeplot(indicator,target,shade=True)
plt.subplot(122)
sns.kdeplot(data_to_eval.H/data_to_eval.H.max(),target,shade=True)
ax.set_xlabel("short-term triads' proportion")
plt.savefig('charts/SW_vs_H.png')
```
